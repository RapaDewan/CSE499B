## Cleaning Data
1. URls
2. Hashtags
3. Mentions
4. Special letters
5. Punctuations

```python
import re

def cleanResume(txt):
    cleanTxt = re.sub(r'http[s]?://\S+', '', txt)  # Matches http or https followed by :// and any non-whitespace characters
    return cleanTxt
```

```python
cleanResume('my website like is this http://helpworls and access it')
```
